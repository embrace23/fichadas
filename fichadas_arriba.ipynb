Creo una fórmula para ser utilizada despues donde se especifica lo siguiente:

- Si el horario está entre 22:00:00 y 23:59:59 traer el menor de esos.
- Si el horario no está dentro de este rango traigo el mínimo.

```python
def horario(series):
    if any(time(22, 0, 0) <= dt.time() <= time(23,59,59) for dt in series):
        return series.max()
    else:
        return series.min()
    
fichadas["Hora"] = pd.to_datetime(fichadas["Hora"])
```

Creo la fichada final en base a la principal donde el indice será "Nombre de Usuario", las columnas todos los valores de la columna original "Día" (es decir todos los días). Los valores serán los horarios mas tempranos por "Nombre de Usuario" en cada "Día".

```python
fichadas_final = fichadas.pivot_table(index="Nombre de Usuario", columns="Día", values="Hora", aggfunc=horario)
fichadas_final
```

Voy recorriendo la fichada final por indice. Si el valor del índice (que sería el nombre) está en algunas de las listas de los sectores que se creó antes, entonces me crea una nueva columna llamada "Sector" por cada fila y le da el valor que le corresponde.

Ejemplo: El primer nombre es "ALVES TEIXEIRA DAVI", va recorriendo lista por lista hasta que coincida el nombre "ALVES TEIXEIRA DAVI" con algún nombre que esté en las listas. En este caso el nombre se encuentra en la lista "front" por lo que a la nueva columna "Sector" se le asignará el valor "Front".

```python
for indice in fichadas_final.index:
    if indice in account:
        fichadas_final.loc[indice, "Sector"] = "Account"
    elif indice in analytics:
        fichadas_final.loc[indice, "Sector"] = "Analytics"
    elif indice in bdx:
        fichadas_final.loc[indice, "Sector"] = "BDX"
    elif indice in billing:
        fichadas_final.loc[indice, "Sector"] = "Billing"
    elif indice in claims:
        fichadas_final.loc[indice, "Sector"] = "Claims"
    elif indice in claims_neg:
        fichadas_final.loc[indice, "Sector"] = "Claims Neg"
    elif indice in comercial:
        fichadas_final.loc[indice, "Sector"] = "Comercial"
    elif indice in front:
        fichadas_final.loc[indice, "Sector"] = "Front"
    elif indice in office:
        fichadas_final.loc[indice, "Sector"] = "Office"
    elif indice in pagos:
        fichadas_final.loc[indice, "Sector"] = "Pagos"
    elif indice in reimbursement:
        fichadas_final.loc[indice, "Sector"] = "Reimbursement"
```
