```python
fichadas["Fecha"] = pd.to_datetime(fichadas["Fecha"]) #Convierto la columna "Día" a formato fecha
```

Creo la fichada final en base a la principal donde el indice será "Name", las columnas todos los valores de la columna original "Fecha" (es decir todos los días). Los valores serán los horarios mas tempranos por "Name" en cada "Fecha".

```python
fichadas_final = fichadas.pivot_table(index="Name", columns="Fecha", values="Hora", aggfunc=min)
fichadas_final
```

Acá traigo las listas del personal por sector. Estar atento a ingreso de personal nuevo o personal que se va. Avisar asi vamos actualizando este listado.

Voy recorriendo la fichada final por indice. Si el valor del índice (que sería el nombre) está en algunas de las listas de los sectores que se creó antes, entonces me crea una nueva columna llamada "Sector" por cada fila y le da el valor que le corresponde.

Ejemplo: Busco el nombre "Lautaro Bustos", va recorriendo lista por lista hasta que coincida el nombre "Lautaro Bustos" con algún nombre que esté en las listas. En este caso el nombre se encuentra en la lista "Office" por lo que a la nueva columna "Sector" se le asignará el valor "Office".

```python
for indice in fichadas_final.index:
    if indice in account:
        fichadas_final.loc[indice, "Sector"] = "Account"
    elif indice in analytics:
        fichadas_final.loc[indice, "Sector"] = "Analytics"
    elif indice in bdx:
        fichadas_final.loc[indice, "Sector"] = "BDX"
    elif indice in billing:
        fichadas_final.loc[indice, "Sector"] = "Billing"
    elif indice in claims:
        fichadas_final.loc[indice, "Sector"] = "Claims"
    elif indice in claims_neg:
        fichadas_final.loc[indice, "Sector"] = "Claims Neg"
    elif indice in comercial:
        fichadas_final.loc[indice, "Sector"] = "Comercial"
    elif indice in office:
        fichadas_final.loc[indice, "Sector"] = "Office"
    elif indice in pagos:
        fichadas_final.loc[indice, "Sector"] = "Pagos"
    elif indice in reimbursement:
        fichadas_final.loc[indice, "Sector"] = "Reimbursement"
    else:
        fichadas_final.loc[indice, "Sector"] = "Otro"
```

En el siguiente código cambio el nombre de las columnas de tipo "fechas" a "texto/string" y reemplazo los "00:00:00" por espacios vacios.

```python
fichadas_final.columns = fichadas_final.columns.astype(str).str.replace(" 00:00:00", "")
```

Busco el mes actual para agregarlo al nombre de los archivos.

```python
# Establecer la configuración regional en español para que el mes se registre en español
locale.setlocale(locale.LC_TIME, 'es_ES.UTF-8')

# Obtener la fecha actual
fecha_actual = datetime.now()

# Obtener el nombre del mes actual
nombre_mes_actual = fecha_actual.strftime("%B").capitalize()
```

Creo los distintos archivos según cada sector y hago que se descarguen.

```python
#ARCHIVO GENERAL
nombre_archivo = f"fichadas Abajo {nombre_mes_actual}.xlsx"
fichadas_final.to_excel(nombre_archivo, index=True)

#ACCOUNT
fichadas_account = fichadas_final[fichadas_final["Sector"] == "Account"]
nombre_account = f"Fichadas Abajo Account {nombre_mes_actual}.xlsx"
fichadas_account.to_excel(nombre_account, index=True)

#Analytics
fichadas_analytics = fichadas_final[fichadas_final["Sector"] == "Analytics"]
nombre_analytics = f"Fichadas Abajo Analytics {nombre_mes_actual}.xlsx"
fichadas_analytics.to_excel(nombre_analytics, index=True)

#BDX
fichadas_BDX = fichadas_final[fichadas_final["Sector"] == "BDX"]
nombre_BDX = f"Fichadas Abajo BDX {nombre_mes_actual}.xlsx"
fichadas_BDX.to_excel(nombre_BDX, index=True)

#Billing
fichadas_Billing = fichadas_final[fichadas_final["Sector"] == "Billing"]
nombre_Billing = f"Fichadas Abajo Billing {nombre_mes_actual}.xlsx"
fichadas_Billing.to_excel(nombre_Billing, index=True)

#Claims
fichadas_Claims = fichadas_final[fichadas_final["Sector"] == "Claims"]
nombre_Claims = f"Fichadas Abajo Claims {nombre_mes_actual}.xlsx"
fichadas_Claims.to_excel(nombre_Claims, index=True)

#Comercial
fichadas_Comercial = fichadas_final[fichadas_final["Sector"] == "Comercial"]
nombre_Comercial = f"Fichadas Abajo Comercial {nombre_mes_actual}.xlsx"
fichadas_Comercial.to_excel(nombre_Comercial, index=True)

#Office
fichadas_Office = fichadas_final[fichadas_final["Sector"] == "Office"]
nombre_Office = f"Fichadas Abajo Office {nombre_mes_actual}.xlsx"
fichadas_Office.to_excel(nombre_Office, index=True)

#Pagos
fichadas_Pagos = fichadas_final[fichadas_final["Sector"] == "Pagos"]
nombre_Pagos = f"Fichadas Abajo Pagos {nombre_mes_actual}.xlsx"
fichadas_Pagos.to_excel(nombre_Pagos, index=True)

#Reimbursement
fichadas_Reimbursement = fichadas_final[fichadas_final["Sector"] == "Reimbursement"]
nombre_Reimbursement = f"Fichadas Abajo Reimbursement {nombre_mes_actual}.xlsx"
fichadas_Reimbursement.to_excel(nombre_Reimbursement, index=True)

#Otro
fichadas_Otro = fichadas_final[fichadas_final["Sector"] == "Otro"]
nombre_Otro = f"Fichadas Abajo Otro {nombre_mes_actual}.xlsx"
fichadas_Otro.to_excel(nombre_Reimbursement, index=True)
```
